## Dataset
### Set
- N = $\{1,2,...,n\}, j \in N$
### Parameters
- $p_j$: processing time of job $j$
- $d_j$: due date of job $j$
- $w_j$: weight of job $j$
- $r_j$: release date of job $j$

```python
# Offline 상황을 가정.

n = 5 # 작업의 개수

N = [_ for _ in range(n)] # 작업 set을 리스트 자료구조를 이용해 0, 1, 2,..., n-1 까지 담아줌
p = [4, 3, 7, 2, 2] # 가공시간 데이터를 리스트 자료구조를 이용해 만들어 줌
d = [5, 6, 8, 8, 17] # 납기 데이터를 리스트 자료구조를 이용해 만들어 줌
w = [5, 4, 3, 2, 1] # 가중치 데이터를 리스트 자료구조를 이용해 만들어 줌
r = [1, 2, 3, 3, 5] # 작업의 release date(도착 시점) 데이터를 리스트 자료구조를 이용해 만들어 줌

M = (sum(p) + sum(r)) * 2 # big M
```

```python
obj_select = 'twt' # 'wct', 'tj', 'twt', 'L_max' 중 하나 선택
```

## 3. Linear ordering variables

### Decision Variables
- $\delta_{j,k}$: 작업 $j$가 작업 $k$보다 앞에서 시작하면 1, 아니면 0인 이진변수, $\delta_{j,k} \in \{0, 1\} \ \forall j \in N, \forall k \in N$
- $S_j$: 작업 $j$의 시작 시점을 의미

```python
# Decision Variable 만들기
delta = {(j, k): cp.Variable(boolean = True) for j in N for k in N}
S = {(j): cp.Variable(nonneg = True) for j in N}
```

### 기본 constraint
- $\delta_{j,k} + \delta_{k,j} = 1 \quad \forall j,k \in N, \ 1 \leq j < k \leq n,$, j<k 가 맞음. 등호 있을 시 infeasible
- $\delta_{j,k} + \delta_{k,l} + \delta_{l,j} \leq 2 \quad \forall j,k,l \in N, \ and \ j \neq k \neq l$

```python
# 제약식 리스트 선언
constraints = []

# 제약식 추가
constraints += [delta[j,k] + delta[k,j] == 1 for j in N for k in N if j < k]
constraints += [delta[j,k] + delta[k,l] + delta[l,j] <= 2 for j in N for k in N for l in N if j != k and k != l and j != l]
```

#### Release time 반영 제약식
$ S_j \geq r_i \delta_{i,j} + \sum_{k < i, k\neq j}p_k(\delta_{i,k} + \delta_{k,j} -1) + \sum_{k \geq i, k \neq j}p_k \delta_{k,j} \quad \forall i, j \in N$

$ \\ \delta_{j,j} = 1, \quad \forall j \in N$
#### 논문의 식 제대로 작동하지 않았음. 새로운 아래 수식으로 대체
$ \\ S_j \geq r_j \quad \forall j \in N$

$ \\ S_k \geq S_j + p_j - M(1-\delta_{j,k}) \quad \forall j \in N, \forall k \in N, j \neq k$

$ \\ S_j \geq S_k + p_k - M\delta_{j,k}  \quad \forall j \in N, \forall k \in N, j \neq k$

```python
# constraints += [S[j] >= r[i]*delta[i,j] + sum(p[k] * (delta[i,k] + delta[k,j] -1) for k in range(0, i) if k != j) + sum(p[k]*delta[k,j] for k in range(i, n) if k != j) for i in N for j in N]
# constraints += [delta[j,j] == 1 for j in N]
constraints += [S[j] >= r[j] for j in N]
constraints += [S[k] >= S[j] + p[j] - M*(1-delta[j,k]) for j in N for k in N if j != k]
constraints += [S[j] >= S[k] + p[k] - M*(delta[j,k]) for j in N for k in N if j != k]
```

$\sum_{k \in N, k \neq j}p_k \delta_{k,j}$ 는 모두 $S_j$로 대체해야 함

#### To minimize the total weighted completion time
$ Minimize \sum_{j \in N}w_j(S_j + p_j)$

```python
if obj_select == 'wct':
    # 목적식
    obj = cp.Minimize(sum(w[j]*(S[j] + p[j]) for j in N))
```

#### To minimize $L_{max}$
$ Minimize \ L_{max}$

$ \\ s.t. \  L_{max} \geq S_j + p_j - d_j \quad \forall j \in N$

```python
if obj_select == "L_max":
    # Lmax 선언
    L_max = cp.Variable(nonneg = True)

    # Lmax 제약식
    constraints += [L_max >= S[j] + p[j] - d[j] for j in N] # C_j 만들지 않고 바로 넣어줌

    # 목적식
    obj = cp.Minimize(L_max)
```

#### To minimize the number of tardy jobs

$ Minimize \ \sum_{j=1}^n U_j$

$ \\ s.t. \  S_j + p_j\leq d_j + MU_j \quad \forall j \in N $

$ \\ U_j \in \{0, 1\} \quad \forall j \in N$

```python
if obj_select == "tj":
    U = {(j): cp.Variable(boolean = True) for j in N} # j가 Tardy job이면 1, 아니면 0인 이진변수

    # Tardy job 제약식
    constraints += [S[j] + p[j] <= d[j] + M*U[j] for j in N]

    # 목적식
    obj = cp.Minimize(sum(U[j] for j in N))
```

#### To minimize the total weighted tardiness

$ Minimize \ \sum_{j=1}^n w_jT_j$

$ \\ s.t. \ T_j \geq S_j + p_j -d_j \quad \forall j \in N$

$ \\ T_j \geq 0 \quad \forall j \in N$

```python
if obj_select == "twt":
    T = {(j): cp.Variable(nonneg = True) for j in N} # j에 대한 Tardiness 연속형 변수, nonneg = True (0 이상)

    # Tardiness 제약식
    constraints += [T[j] >= S[j] + p[j] - d[j] for j in N] # C_j 없이 바로 넣어줌

    # 목적식
    obj = cp.Minimize(sum(w[j]*T[j] for j in N))
```

```python
prob = cp.Problem(obj, constraints)
prob.solve()
```

```python
print("최적해 상태:", prob.status)
print("최적값:", prob.value)
```
